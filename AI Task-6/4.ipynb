```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import Birch 

data = pd.read_csv('./Mall_Customers.csv')
data
```

```python
X = data[["Annual Income (k$)","Spending Score (1-100)"]]
X
```

```python
plt.scatter(data["Annual Income (k$)"], data["Spending Score (1-100)"], s = 30, c = 'r')
plt.show()
```

```python
model = Birch(branching_factor =30, n_clusters = 5, threshold = 2.5) 
model.fit(X)
pred = model.predict(X)
plt.scatter(data["Annual Income (k$)"], data["Spending Score (1-100)"], c = pred, cmap = 'rainbow', alpha = 1, edgecolors = 'b') 
plt.show()
```
