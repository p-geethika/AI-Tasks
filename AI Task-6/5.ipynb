```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import DataFrame
import seaborn as sns

data = pd.read_csv('./CC GENERAL.csv')
data = data.drop('CUST_ID', axis = 1)
data.fillna(method ='ffill', inplace = True)
data
```

```python
from sklearn.cluster import MeanShift
from mpl_toolkits.mplot3d import Axes3D

X = data.iloc[:, [1,3,4]].values
ms = MeanShift(bandwidth=12)
y_ms = ms.fit_predict(X)
fig = plt.figure(figsize=(14, 20))
plt.subplot(211, projection='3d')
plt.scatter(X[:, 0], X[:, 1],X[:,2], c=y_ms, cmap='rainbow')
centers = ms.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1],centers[:,2], c='black')
plt.title('Clusters using MeanShift', fontsize=25)

```
